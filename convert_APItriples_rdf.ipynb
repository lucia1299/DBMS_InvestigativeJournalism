```python
new_vocabulary_prefixes = {
    'rdf': ['http://www.w3.org/1999/02/22-rdf-syntax-ns#type'],
    'bibo': ['http://purl.org/ontology/bibo/citedBy'],
    'owl': ['http://www.w3.org/2002/07/owl/sameAs']
}

new_prefixed_names = {
    'rdf:type': 'http://www.w3.org/1999/02/22-rdf-syntax-ns#type',
    'bibo:citedBy': 'http://purl.org/ontology/bibo/citedBy',
    'owl:sameAs': 'http://www.w3.org/2002/07/owl/sameAs'
}

# Updating the dictionary
RDFprefixes['vocabulary_prefixes'].update(new_vocabulary_prefixes)
RDFprefixes['prefixed_names'].update(new_prefixed_names)

RDFprefixes
```

```python
from urllib.parse import urlparse

def extract_base_uri(uri):
    parsed_uri = urlparse(uri)
    base_uri = f"{parsed_uri.scheme}://{parsed_uri.netloc}{parsed_uri.path}"
    if '#' in uri:
        return base_uri.split('#')[0] + '#'
    elif '/' in uri:
        return base_uri.rsplit('/', 1)[0] + '/'
    return base_uri

def process_vocabulary_prefixes(vocab_dict):
    base_uri_dict = {}
    for prefix, uris in vocab_dict.items():
        if uris:
            base_uri = extract_base_uri(uris[0])
            base_uri_dict[prefix] = base_uri
    return base_uri_dict

processed_vocabulary_prefixes = process_vocabulary_prefixes(RDFprefixes['vocabulary_prefixes'])

processed_vocabulary_prefixes
```

```python
for prefix, uri in processed_vocabulary_prefixes.items():
  print(uri)
```

```python
from rdflib import Graph, URIRef, Literal, Namespace
from urllib.parse import quote

g = Graph()

base_namespace = Namespace("http://newsorganiser.org/resource/")

for prefix, uri in processed_vocabulary_prefixes.items():
    g.bind(prefix, uri)

def str_to_uri(base_namespace, s):
    return URIRef(base_namespace + quote(s))

for triple in merged_triples:
    subject = str_to_uri(base_namespace, triple['subject'])
    predicate = URIRef(triple['predicate'])

    # Convert object to URI using base namespace if it's a literal
    if isinstance(triple['object'], (int, float)):
        obj = Literal(triple['object'])
    elif triple['object'].startswith('http'):
        obj = URIRef(triple['object'])
    else:
        obj = str_to_uri(base_namespace, triple['object'])

    g.add((subject, predicate, obj))

# Serialize the graph to Turtle format
rdf_data = g.serialize(format='turtle')
print(rdf_data)

# ns1 appears because prefix rdarel is deprecated
# how is the score calculated of each vocab term in lov?
# that prefix had the highest score in category 'employer' and still it was not recognised by rdflib
# is it because it is deprecated?
```

```python
import pandas as pd
import re
from rdflib import Graph, URIRef, Literal, RDF, Namespace
from rdflib.namespace import NamespaceManager

# Create an RDF graph
g = Graph()

# Base namespace for node IDs
base_ns = "http://news_ontology.org/"

# Predefined prefixes
namespace_prefixes = {
    base_ns: "NewsOnt:"
}

import re

def create_prefix(namespace):
    # Normalize the namespace to remove the fragment after the final '/' or '#'
    if '#' in namespace:
        parts = namespace.split('#')
        namespace_base = parts[0]
    else:
        namespace_base = namespace.rstrip('/')

    # Extract the last segment before the final '/' or '#'
    prefix = namespace_base.split('/')[-1].replace(':', '_')

    # Ensure the prefix ends with a single colon
    prefix = prefix.rstrip(':') + ':'

    # Prefix must start with a letter and can contain letters, digits, and underscores
    if not re.match(r'^[A-Za-z][A-Za-z0-9_]*:$', prefix):
        prefix = 'ns' + str(len(namespace_prefixes) + 1) + ':'

    return prefix

# Function to bind namespace automatically if not already bound
def ensure_namespace_bound(uri):
    # Check if the URI is an RDF literal or has a full namespace URI
    if isinstance(uri, URIRef):
        namespace = create_prefix(uri)
        if namespace not in namespace_prefixes:
            prefix = create_prefix(namespace)
            namespace_prefixes[namespace] = prefix
            g.namespace_manager.bind(prefix.rstrip(':'), Namespace(namespace), override=True)
    return

# Function to determine if a URI is a Wikidata URI
def is_wikidata_uri(uri):
    return str(uri).startswith('https://www.wikidata.org/wiki/')

# Add nodes to the RDF graph
for _, row in nodes_df.iterrows():
    node_id = row['Id']
    node_label = row['Label']

    node_uri = URIRef(base_ns + str(node_id))

    # Ensure the base namespace is bound
    ensure_namespace_bound(node_uri)

    # Determine if the label is a URI or a Literal
    if node_label.startswith('http://') or node_label.startswith('https://'):
        label = URIRef(node_label)  # Treat as URI
        # Ensure full URI for Wikidata links
        if is_wikidata_uri(label):
            g.add((node_uri, RDF.type, label))
        else:
            ensure_namespace_bound(label)
            g.add((node_uri, RDF.type, label))
    else:
        label = Literal(node_label)  # Treat as Literal
        g.add((node_uri, RDF.type, label))

# Add edges to the RDF graph
for _, row in edges_df.iterrows():
    source_uri = URIRef(base_ns + str(row['source']))
    target_uri = URIRef(base_ns + str(row['target']))
    predicate_uri = URIRef(row['label'])

    # Ensure namespaces are bound for source, target, and predicate URIs
    if not is_wikidata_uri(source_uri):
        ensure_namespace_bound(source_uri)
    if not is_wikidata_uri(target_uri):
        ensure_namespace_bound(target_uri)
    if not is_wikidata_uri(predicate_uri):
        ensure_namespace_bound(predicate_uri)

    # Add edge triples to the graph
    g.add((source_uri, predicate_uri, target_uri))

# Serialize the RDF graph to a string in the Turtle format
rdf_data = g.serialize(format='turtle')

print(rdf_data)
```

```python
# Save to a file
with open(turtle_file, "w") as file:
    file.write(rdf_data)
```
